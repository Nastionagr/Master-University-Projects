```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
df_0 = pd.read_csv('big_results_0')
df_1 = pd.read_csv('big_results_1')
df_2 = pd.read_csv('big_results_2')
df_3 = pd.read_csv('big_results_3')
df_4 = pd.read_csv('big_results_4')
df_5 = pd.read_csv('big_results_5')
df_6 = pd.read_csv('big_results_6')
df_7 = pd.read_csv('big_results_7')
df_8 = pd.read_csv('big_results_8')
df_9 = pd.read_csv('big_results_9')
df_10 = pd.read_csv('big_results_10')
df_11 = pd.read_csv('big_results_11')
df_12 = pd.read_csv('big_results_12')
df_13 = pd.read_csv('big_results_13')
df_14 = pd.read_csv('big_results_14')
df_15 = pd.read_csv('big_results_15')
```

```python
def add_info(df, complexity):
    df["Complexity"] = complexity
    df.drop(columns=["Unnamed: 0", "Task"], inplace=True)

_ = [add_info(df, "M") for df in [df_0, df_4, df_8,  df_12]]
_ = [add_info(df, "N") for df in [df_1, df_5, df_9,  df_13]]
_ = [add_info(df, "Q") for df in [df_2, df_6, df_10, df_14]]
_ = [add_info(df, "R") for df in [df_3, df_7, df_11, df_15]]
```

```python
df = pd.concat([df_0, df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12, df_13, df_14, df_15])
# some of the durations are not legal as my computer was in the sleeping mode and the calculations stopped for that time
df = df[df["Duration"] < 10000]
df
```

```python
df_ss = df[df.isSolutionFound == True]
df_s  = df[(df.isSolutionFound == False) & (df.bestWeight > 0)]
df_u  = df[(df.isSolutionFound == False) & (df.bestWeight == 0)]

print("Number of successful solutions:\t\t\t" + str(np.round(len(df_ss) / len(df) * 100, 3)) + " %")
print("Number of solved, but not int the best way:\t" + str(np.round(len(df_s) / len(df) * 100, 3)) + " %")
print("Number of unsolved:\t\t\t\t" + str(np.round(len(df_u) / len(df) * 100, 3)) + " %")
```

```python
plt.subplots(1,2,figsize=(15,5))

df["doesSolutionExist"] = np.where(df['bestWeight']!=0, True, False)

plt.subplot(1,2,1); sns.countplot(x = "Complexity", hue = "isSolutionFound" , data = df);
plt.subplot(1,2,2); sns.countplot(x = "Complexity", hue = "doesSolutionExist" , data = df);

df.drop(columns=["doesSolutionExist"], inplace=True)
```

```python
df_not_opt = df[(df["bestWeight"] != 0) & (df["isSolutionFound"] == False)].reset_index()
df_not_opt["Error"] = (df_not_opt["optWeight"] - df_not_opt["bestWeight"]) / df_not_opt["bestWeight"] * 100
df_not_opt.sort_values("Error", ascending=False).head(10)
```

```python
plt.subplots(1,2,figsize=(15,5))

plt.subplot(1,2,1); sns.boxplot(x='Error', data=df_not_opt[df_not_opt["Error"]<500]);
plt.subplot(1,2,2); df_not_opt[df_not_opt["Error"]<500]["Error"].hist();
```
