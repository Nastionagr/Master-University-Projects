```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results')
```

```python
_ = df.sort_values(by=['avgEndGeneration', 'avgDuration', 'numFoundedSolutions'])["Parameters"].head(5).apply(lambda x: print(x))
```

```python
df.sort_values(by=['avgEndGeneration', 'avgDuration', 'numFoundedSolutions']).head(5)
```

```python
df["populationSize"] = df.Parameters.apply(lambda x: int(x.split()[0].replace("[","").replace("'", "")))
df["parentSelectionMethod"] = df.Parameters.apply(lambda x: x.split()[1].replace("'", ""))
df["crossoverMethod"] = df.Parameters.apply(lambda x: x.split()[2].replace("'", ""))
df.loc[df['crossoverMethod'] == "Wheel", 'crossoverMethod'] = df[df["crossoverMethod"] == "Wheel"]["Parameters"].apply(lambda x: x.split()[3].replace("'", ""))
df["crossoverRate"] = df.Parameters.apply(lambda x: float(x.split()[-3].replace("'", "")))
df["minMutationRate"] = df.Parameters.apply(lambda x: float(x.split()[-2].replace("'", "")))
df["maxMutationRate"] = df.Parameters.apply(lambda x: float(x.split()[-1].replace("]","").replace("'", "")))
# df["isMutated"] = df.Parameters.apply(lambda x: float(x.split()[-1].replace("]","").replace("'", ""))>0)
df = df.drop(columns=["Unnamed: 0", "Task", "Parameters"])
df
```

```python
plt.pie(df['numFoundedSolutions'].value_counts(), labels=[3, 2, 1, 0], autopct='%.0f%%')
```

```python
plt.subplots(1,2,figsize=(15,5))

plt.subplot(1,2,2); sns.countplot(x = "numFoundedSolutions", hue = "maxMutationRate" , data = df);
plt.subplot(1,2,1); sns.countplot(x = "numFoundedSolutions", hue = "minMutationRate" , data = df);
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x = "numFoundedSolutions", hue = "crossoverRate" , data = df);
```

```python
sns.displot(x = "avgEndGeneration", hue = "crossoverRate" , data = df);
```

```python
df.groupby(["parentSelectionMethod", "crossoverMethod"]).numFoundedSolutions.sum()
```

```python
df.groupby(["parentSelectionMethod", "crossoverMethod"]).numFoundedSolutions.value_counts().plot(kind='bar')
```

```python
sns.countplot(x = "numFoundedSolutions", hue = "populationSize" , data = df);
```
